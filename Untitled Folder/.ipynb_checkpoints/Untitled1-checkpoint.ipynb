```python
import matplotlib.pyplot as plt
location_list = sales_data.groupby('Location')
location_list
```

```python
location = [x for x,y in location_list]
location
plt.bar(location, sales_data.groupby('Location').sum(numeric_only=True)['Total'])
```

```python
plt.pie(sales_data.groupby('Location').sum(numeric_only=True)['Total'], autopct = '%1.1f%%', labels = location)
```

```python
location_sales = sales_data.groupby(['Gender', 'Location']).count()['Invoice ID']
unstacked = location_sales.unstack(level = 0)
unstacked.plot(kind='bar')
```

```python
sales_data['Day']=pd.to_datetime(sales_data['Date']).dt.day
sales_data.head(10)
```

```python
day_sales = sales_data.groupby('Day').sum(numeric_only = True)['Total']
day_sales.plot()
```

```python
members = sales_data.groupby(['Member', 'Location']).count()['Invoice ID']
unstacked_members = members.unstack(level = 0)
unstacked_members.plot(kind = 'bar')
```

```python
rating = sales_data.groupby('Location').mean(numeric_only = True)['Rating']
rating.plot(kind = 'bar')
```

```python
#which type of customer spends more member or non-member

members_spends = sales_data.groupby('Member').sum(numeric_only = True)['Total']
members_spends.plot(kind = 'bar', xlabel = 'Member Status').set_xticklabels(['Non-Member', 'Member'])
```

```python
#which product line sells more

product_line = sales_data.groupby('Category')['Total'].sum()
product_line.plot(kind = 'bar')
```

```python
#which product line is popular among men and women

popular_product = sales_data.groupby([ 'Gender', 'Category']).sum(numeric_only = True)['Total']
popular_product.plot(kind = 'bar')
```

```python
sales_data['Month'] = pd.to_datetime(sales_data['Date']).dt.month
sales_data.head(10)
```

```python
month_total = sales_data.groupby('Month')['Total'].sum()
total_unstaked = month_total.unstack(level = 0)
total_unstaked.plot(kind = 'bar')
```
